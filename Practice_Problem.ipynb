**PRACTICE PROBLEM**

Design an algorithm and the corresponding flowchat for adding the test scores as given if the number is even:

26,49,98,87,62,75

```python
scores = [26, 49, 98, 87, 62, 75]
sum = 0

for score in scores:
    if score % 2 == 0:
        sum += score

print("Sum of even scores:", sum)
```
